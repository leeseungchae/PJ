```python
from gensim.models import Word2Vec
import pandas as pd

review_word = pd.read_csv('./datasets/movie_review_2018_2022.csv')
review_word.info()
```

```python
cleaned_token_review = list(review_word['cleaned_sentences'])
# print(cleaned_token_review[0])
cleaned_tokens = []
for sentence in cleaned_token_review:
    token = sentence.split()  #default 공백
    cleaned_tokens.append(token)
print(cleaned_tokens[0])
```

```python
embedding_model = Word2Vec(cleaned_tokens, vector_size=100 , window=4 , min_count=20, workers=4, epochs= 100 , sg=1)
embedding_model.save('./models/word2vecModel.model') #문장을 4개씩 자름 , 최소빈도수 안나오면 학습x , 코어개수 , 알고리즘 종류
# print(embedding_model.wv.vocab.keys())
print(embedding_model.wv.index_to_key)
```
